**IMPORT MODULES**

```python
# Importing Modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**IMPORTING DATAS**

```python
# Importing Data if you have data
df=pd.read_csv('https://raw.githubusercontent.com/jsdhami/Python-For-Research/main/Data/nepal-earthquake-severity-index-latest.csv')

df = pd.DataFrame({'District':df['DISTRICT'], 'Housing':df['Housing'], 'Poverty':df['Poverty'], 'Vulnerability':df['Vulnerability']})
df.head()
print(len(df))
```

```python
df = df.tail(8)
df.head()
print(df.to_string())
```

**CREAT DATA FRAME**

```python
# Create Data
x=df['District'].astype(str)
y=df['Housing']
z=df['Poverty']
v=df['Vulnerability']
```

**A. LINE PLOTING**

```python
# Line Ploting For All
# Creating Fig
fig, ax = plt.subplots()
fig.set_size_inches(15, 5)

ax.plot(x, y, 'o-', label='Housing',  color='blue')
ax.plot(x, z, 'o--', label='Poverty', color='black')
ax.plot(x, v, 'o-', label='Vulnerability', color='red')

# label
ax.set_xlabel('District', color='green')
ax.set_ylabel('Values', color='green')
ax.set_title('Nepal Earthquake Severity Index',
             color='green', fontsize=12, weight=20)

#Stick/Ticks Label
ax.set_xticks(x)
ax.set_ylim(1, 12)

ax.legend()

plt.grid(True)

plt.show()

```
